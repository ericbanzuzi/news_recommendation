```python
import py7zr

def unzip_7z_file(file_path, output_dir):
    with py7zr.SevenZipFile(file_path, mode='r') as z:
        z.extractall(path=output_dir)
    print(f"Extracted all files to {output_dir}")

# Example usage
file_path = '/content/drive/MyDrive/Colab Notebooks/FOX.7z'
output_dir = '/content/drive/MyDrive/OUTput'
unzip_7z_file(file_path, output_dir)
```

**Specify the folders to use in summary addition:**

```python
input_folder_path = '/content/sun'
output_folder_path = '/content/drive/MyDrive/dataset/sun_final'
```

```python
import torch
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM
import transformers


transformers.utils.HTTP_TIMEOUT = 100
device = 0 if torch.cuda.is_available() else -1
model_name = "sshleifer/distilbart-cnn-12-3"
summarizer = pipeline("summarization", model=model_name, device=device)
```

```python
counter=0
filenames = sorted(os.listdir(input_folder_path))
for filename in filenames:
        counter+=1
        if counter==500:
          print("Processed files "+str(counter))
        file_path = os.path.join(input_folder_path, filename)
        with open(file_path, 'r') as file:
            data = json.load(file)
            content = data['content']
            if(len(content)>500):
              content=content[:500]
            content2=[content]
            summarized_text = summarizer(content2, max_length=130, min_length=50, do_sample=False)[0]
            summarized_text = summarized_text['summary_text']
            data['summary'] = summarized_text
            output_file_path = os.path.join(output_folder_path, filename)
            with open(output_file_path, 'w') as output_file:
                json.dump(data, output_file, ensure_ascii=False, indent=4)
```

```python
import os

def count_files(directory):
    """Count all files in the directory and subdirectories."""
    total_files = 0
    for root, dirs, files in os.walk(directory):
        total_files += len(files)
    return total_files

# Specify the path to your directory
directory_path = '/content/drive/MyDrive/dataset/sun_final'

# Get the count of files
file_count = count_files(directory_path)
print(f"Total number of files in '{directory_path}': {file_count}")
```
